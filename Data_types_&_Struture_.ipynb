2. Explain the difference between mutable and immutable data types with examples?

Ans: The difference between mutable and immutable data types lies in whether their values can be changed after they are created.

Mutable Data Types:
Definition: Mutable data types can be modified after they are created.
Examples:
Lists: You can add, remove, or change elements in a list.
python





```python
#Lists: You can add, remove, or change elements in a list.
my_list = [1, 2, 3]
my_list[0] = 10
my_list.append(4)
print(my_list)
```

Immutable Data Types:
Definition: Immutable data types cannot be changed once they are created.

```python
#Tuples: You cannot modify, add, or remove elements from a tuple.
my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # This would raise an error
```

```python
#Strings: Once a string is created, its characters cannot be changed.
my_str = "hello"
# my_str[0] = "H"  # This would raise an error
```

Key Difference:
Mutable types can be modified after creation, while immutable types cannot be altered once they're set.

4. Describe how dictionaries store data?

Ans:Dictionaries in Python store data as key-value pairs. Each key is unique, and it maps to a specific value. This structure allows for fast lookups, as data can be accessed using the key rather than iterating through elements.

How Dictionaries Store Data:
Key-Value Pair: A dictionary consists of pairs, where the key is used to access the value.
Keys must be immutable types (e.g., strings, numbers, tuples).
Values can be of any data type, including mutable types like lists or other dictionaries.

Example:

my_dict = {'name': 'Alice', 'age': 25, 'city': 'New York'}

In this example:
'name', 'age', and 'city' are the keys.
'Alice', 25, and 'New York' are the values associated with each key.

Key Features:

Unordered: In older Python versions, dictionaries do not guarantee the order of key-value pairs. Starting from Python 3.7, dictionaries maintain the insertion order.
Efficient Lookup: You can access values quickly using the key.

For example, my_dict['name'] will return 'Alice'.

Example of accessing data:

print(my_dict['name'])  # Output: Alice
Modifying a Dictionary:
You can change the value associated with an existing key, add new key-value pairs, or remove them.

my_dict['age'] = 26  # Modify value
my_dict['job'] = 'Engineer'  # Add new key-value pair
del my_dict['city']  # Remove key-value pair







5. Why might you use a set instead of a list in Python?

Ans: use a set instead of a list in Python for the following reasons:

Uniqueness:

Sets automatically store only unique elements. If you try to add a duplicate item, it will be ignored.
Lists allow duplicates.
Example:


my_set = {1, 2, 3, 3}  # Output: {1, 2, 3}
my_list = [1, 2, 3, 3]  # Output: [1, 2, 3, 3]
Faster Membership Testing:

Sets provide faster membership testing (checking if an item exists) compared to lists. This is because sets are implemented using a hash table, making lookups average O(1) time complexity.
Lists require O(n) time complexity for membership testing, where n is the size of the list.

Example:

my_set = {1, 2, 3}
2 in my_set  # Faster, O(1) lookup

my_list = [1, 2, 3]
2 in my_list  # Slower, O(n) lookup
No Ordering:

Sets are unordered, meaning the elements do not have a fixed position.
Lists are ordered, and their elements have a specific index.
If the order of elements is not important, sets are a good choice. If order matters, a list would be more appropriate.

Set Operations:

Sets support mathematical operations like union, intersection, and difference, which are efficient.
These operations are not directly supported by lists.

Example:

set1 = {1, 2, 3}
set2 = {2, 3, 4}
union = set1 | set2  # {1, 2, 3, 4}
intersection = set1 & set2  # {2, 3}
difference = set1 - set2  # {1}
Use Cases for Sets:
When you need to ensure that the collection only contains unique elements.
When fast membership testing is crucial.
When performing set operations like union, intersection, or difference.
Example where a set is useful:

# Removing duplicates from a list
my_list = [1, 2, 2, 3, 4, 4]
my_set = set(my_list)  # {1, 2, 3, 4}






6. What is a string in Python, and how is it different from a list?

Ans: A string in Python is a sequence of characters enclosed in single (') or double (") quotes. It is an immutable data type, meaning once a string is created, its content cannot be changed.

Key Differences between a String and a List in Python:
Mutability:

String: Immutable (cannot be modified after creation). You cannot change individual characters in a string.
List: Mutable (can be modified). You can change, add, or remove elements in a list.
Data Type:

String: Contains a sequence of characters (letters, numbers, or symbols).
List: Can store a sequence of any type of objects, including integers, strings, other lists, or even mixed types.
Syntax:

String: Defined with single or double quotes, e.g., 'hello' or "world".
List: Defined with square brackets, e.g., [1, 2, 3].
Operations:

String: Strings support operations like concatenation, slicing, and indexing but do not allow modification of individual characters.
List: Lists support more operations like adding/removing elements (append(), remove()), modifying elements, and changing the list's length.
Examples:
String Example:


my_string = "hello"
# You cannot do this:
# my_string[0] = "H"  # This will raise an error because strings are immutable
List Example:



my_list = [1, 2, 3]
my_list[0] = 10  # Modifying the list
print(my_list)  # Output: [10, 2, 3]
Key Points:
String: A sequence of characters, immutable.
List: A sequence of elements, mutable, and can store any type of data.
